```python
from deepface import DeepFace
import cv2
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('images/monalisa.jpeg')
plt.imshow(img1[:,:,::-1])
plt.show
```

```python
result = DeepFace.analyze(img1, actions=['emotion'])
```

```python
print(result)
```
